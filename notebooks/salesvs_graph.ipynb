```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../dataset/advertising.csv')
df.head(5)
```

```python
plt.figure(figsize=(8,5))
plt.grid()
plt.scatter(x=df['Sales'],y=df['TV'],color='red')
plt.xlabel("Sales")
plt.ylabel("TV Expness")
plt.show()

#Radio
plt.figure(figsize=(8,5))
plt.grid()
plt.scatter(x=df['Sales'],y=df['Radio'],color='green')
plt.xlabel("Sales")
plt.ylabel("Radio")
plt.show()

#Newspaper
plt.figure(figsize=(8,5))
plt.grid()
plt.scatter(x=df['Sales'],y=df['Newspaper'],color='yellow')
plt.xlabel("Sales")
plt.ylabel("Newspaper")

plt.show()
```
